```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df
```

```python
X = df.iloc[:, 0:2].values
y = df.iloc[:, -1].values
```

```python
y_binary = np.where(y == "setosa", 1.0, 0.0).astype(np.float32)
y_binary
```

```python
X = (X - X.mean(axis=0))/X.std(axis=0)
X
```

```python
X_tf = tf.constant(X, dtype = tf.float32)
X_tf
```

```python
W = tf.constant([[1.0],
                [-1.0]], dtype = tf.float32)
b = tf.constant([0.2], dtype = tf.float32)
```

```python
z = tf.matmul(X_tf, W) +b
z
```

```python
y_pred_step = tf.cast(z >= 0, tf.int32)
```

```python
y_pred_sigmoid = tf.math.sigmoid(z)
```

```python
print("First 10 True labels(Setosa=1, Others=0): ",y_binary[:10])
print("First 10 step outputs: ",y_pred_step.numpy().flatten()[55:-1])
print("First 10 sigmoid probabilities:",y_pred_sigmoid.numpy().flatten()[:10])
```
